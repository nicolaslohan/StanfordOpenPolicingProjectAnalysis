```python
import pandas as pd
import numpy as np
```

## Preparando dados para análise

```python
police = pd.read_csv('police.csv', sep=',')
weather = pd.read_csv('weather.csv', encoding='utf_8',sep=',')
```

```python
size = police.shape
print(f'Há {size[0]} linhas e {size[1]} colunas')
```

```python
colunas = police.columns.tolist()
for i in colunas:
    print(f'Há {police[i].isna().sum()} valores vazios na coluna "{i}"')
```

```python
police.drop(['county_name', 'state'], axis=1, inplace=True)
```

```python
police = police.dropna(subset=['driver_gender'], axis=0)
```

```python
merge = police.assign(stop_date = police.stop_date + " " + police.stop_time + ":00")
```

```python
merge
```

```python
merge.rename(columns = {'stop_date': 'stop_datetime'}, inplace=True)
```

```python
merge
```

```python
merge.drop(['stop_time'], axis=1, inplace=True)
```

```python
merge
```

```python
merge['stop_datetime'] = pd.to_datetime(merge['stop_datetime'])
```

```python
merge.dtypes
```

```python
merge.set_index('stop_datetime', inplace=True)
```

```python
merge
```

## Análise de gênero sobre as infrações

```python
violations = pd.DataFrame(merge.groupby(['violation']).agg(n_occurrences = ('violation', 'count'))
violations.reset_index(inplace=True)
print(violations.max())
print("\n")
print(violations.min())
```

```python
viol_gender = merge[['violation', 'driver_gender']]
gen = viol_gender['driver_gender'].unique
```

```python
gp = viol_gender.groupby(['violation', 'driver_gender']).agg(num = ('driver_gender', 'count'))
```

```python
gp.reset_index(level=[1])
```

```python
#gp.set_index('num')
gp.pivot_table(index='violation',columns='driver_gender',values='num',aggfunc='count')
```
