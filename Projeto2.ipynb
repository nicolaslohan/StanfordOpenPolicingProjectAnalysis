## Preparando dados para análise

```python
police = pd.read_csv('police.csv', sep=',')
weather = pd.read_csv('weather.csv', encoding='utf_8',sep=',')
```

```python
size = police.shape
print(f'Há {size[0]} linhas e {size[1]} colunas')
```

```python
colunas = police.columns.tolist()
for i in colunas:
    print(f'Há {police[i].isna().sum()} valores vazios na coluna "{i}"')
```
