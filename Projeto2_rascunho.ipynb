## Análise de gênero sobre as infrações

```python
violations = pd.DataFrame(merge.groupby(['violation']).agg(n_occurrences = ('violation', 'count')))
violations.reset_index(inplace=True)
print(violations.max())
print("\n")
print(violations.min())
```

```python
vg = merge[['violation', 'driver_gender']]
```

```python
table = vg.loc[merge['driver_gender'] == 'F']
table
```

```python
freqf = pd.DataFrame(vg[vg['driver_gender']=='F'].value_counts(normalize=True)*100)
freqf.rename(columns={0:'rel_freq_female'}, inplace=True)
freqf['absolutef_female'] = vg[vg['driver_gender']=='F'].value_counts()
freqf['rel_freq_female'] = freqf['rel_freq_female'].round(2)
freqf.reset_index(inplace=True)
freqf.drop('driver_gender', axis=1, inplace=True)
freqf
```

```python
freqm = pd.DataFrame(vg[vg['driver_gender']=='M'].value_counts(normalize=True)*100)
freqm.rename(columns={0:'rel_freq_male'}, inplace=True)
freqm['absolutef_male'] = vg[vg['driver_gender']=='M'].value_counts()
freqm['rel_freq_male'] = freqm['rel_freq_male'].round(2)
freqm.reset_index(inplace=True)
freqm.drop('driver_gender', axis=1, inplace=True)
freqm
```

```python
general = freqm.merge(freqf, how='inner', right_on='violation', left_on='violation')
general.set_index('violation', inplace=True)
general
```

```python
graph_abf = general[['absolutef_male', 'absolutef_female']]
graph_abf.plot(kind='bar', title='Absolute Frequence of Violations by Gender')
```

```python
graph_relf = general[['rel_freq_male', 'rel_freq_female']]
graph_relf.plot(kind='bar', title='Relative Frequence of Violations by Gender')
```
